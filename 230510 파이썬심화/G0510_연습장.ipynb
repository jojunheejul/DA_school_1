```python
#수집된 서울의 기온 데이터를 히스토그램으로 표시하시오.
#색은 빨간색
#구간은 100개로 나누시오.
import os
os.chdir('/content/')

import csv

f = open('/content/CSV_example2_data.csv', 'r') #csv를 리드 모드로 불러옴
data = csv.reader(f)
next(data)

result = [] 

for row in data: 
  result.append(float(row[-1]))      
f.close()

import matplotlib.pyplot as plt

#plt.figure(figsize = (10, 2)) #그래프의 가로길이를 10인치 세로를 2인치로 설정
plt.hist(result,color = 'g', bins = 10) #데이터의 분포를 나타냄 bins는 구간을 100으로 현재 나눈 상태
plt.show()
```

```python
import os
os.chdir('/content/')

import csv

f = open('/content/CSV_example2_data.csv', 'r') #csv를 리드 모드로 불러옴
data = csv.reader(f)
next(data)

result_04 = [] 
result_05 = [] 

for row in data:    
  month = row[0].split('-')[1]
  if month == '04':
    result_04.append(float(row[-1])) 
  if month == '05': 
    result_05.append(float(row[-1]))       
f.close()

import matplotlib.pyplot as plt

#plt.figure(figsize = (10, 2)) #그래프의 가로길이를 10인치 세로를 2인치로 설정
plt.hist(result_04,color = 'b', bins = 10) #데이터의 분포를 나타냄 bins는 구간을 100으로 현재 나눈 상태
plt.hist(result_05,color = 'g', bins = 10) #데이터의 분포를 나타냄 bins는 구간을 100으로 현재 나눈 상태
plt.show()
```

```python
import csv
f = open('/content/dataset_example3_new (1).csv')
data = csv.reader(f)

female = [] #여성 인구정보를 담을 리스트
male = [] #남성 인구정보를 담을 리스트

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[16:27]:#1에서 100까지
      i = i.replace(",", "")
      male.append(int(i)) #남성데이터는 15~27
    for i in row[29:]:
      i = i.replace(",", "")
      female.append(int(i)) #여성데이터는 30~번까지

import matplotlib.pyplot as plt

plt.bar(range(11), male)
plt.bar(range(11), female)
plt.legend()
plt.show()
f.close()
```
