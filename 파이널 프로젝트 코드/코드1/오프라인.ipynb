```python
# 월별로 그룹화하고 결제금액을 합산합니다.
monthly_payment_sum = off_df.groupby(off_df['결제일'].dt.to_period('M'))['합계'].sum()
monthly_payment_sum
```

```python
# 월별 결제금액 합계를 DataFrame으로 변환합니다.
monthly_payment_df = monthly_payment_sum.to_frame(name='월별 결제금액 합계')

monthly_payment_df
```

```python
total_sum = monthly_payment_df.sum()

# 결과 출력
print("전체 합계:", total_sum)
```

```python
import matplotlib.ticker as ticker

# 월별 결제일 데이터를 datetime 형식으로 변환
# monthly_payment_df.index = monthly_payment_df.index.to_timestamp()

# 그래프를 그리기 위해 시각화 설정
plt.figure(figsize=(12, 6))  # 그래프 크기 조정
plt.plot(monthly_payment_df.index.strftime('%Y-%m'), monthly_payment_df['월별 결제금액 합계'], marker='o', linestyle='-')
plt.xlabel('월')
plt.ylabel('결제금액 합계')
plt.title('월별 결제금액 추세')

# x 축에 모든 월을 표시하는 설정
ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(base=1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=1))

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
