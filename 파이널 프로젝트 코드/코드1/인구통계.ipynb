```python
import pandas as pd

people_df = pd.concat([people_1, people_2, people_3])

gender_age_payment_sum1 = people_1.groupby(['성별', '나이'])['결제금액'].sum()
gender_age_payment_sum2 = people_2.groupby(['성별', '나이'])['결제금액'].sum()
gender_age_payment_sum3 = people_3.groupby(['성별', '나이'])['결제금액'].sum()
gender_age_payment_sum4 = people_df.groupby(['성별', '나이'])['결제금액'].sum()

result_df = pd.concat([gender_age_payment_sum1, gender_age_payment_sum2, gender_age_payment_sum3, gender_age_payment_sum4], axis=1)
result_df.columns = ['people_1 결제금액', 'people_2 결제금액', 'people_3 결제금액', '전체 결제금액']

result_df['전체 결제금액 대비 비율'] = result_df['전체 결제금액'] / result_df['전체 결제금액'].sum()

result_df['전체 결제금액 대비 비율'] = result_df['전체 결제금액 대비 비율'].apply(lambda x: '{:.1%}'.format(x))

result_df_sorted = result_df.sort_values(by='전체 결제금액', ascending=False)

result_df_sorted
```

```python
plt.figure(figsize=(12, 6))
result_df_sorted['전체 결제금액'].plot(kind='bar', color='skyblue', edgecolor='black', width=0.6)
plt.xlabel('성별, 나이 그룹')
plt.ylabel('전체 결제금액')
plt.title('성별과 나이 그룹별 전체 결제금액')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
