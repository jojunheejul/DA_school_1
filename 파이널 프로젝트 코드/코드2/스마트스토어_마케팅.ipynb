```python
import koreanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_excel('/content/3. 스마트스토어_상품_마케팅채널_2022-01-01_2022-06-30.xlsx')
df_2 = pd.read_excel('/content/3. 스마트스토어_상품_마케팅채널_2022-07-01_2022-12-31.xlsx')
df_3 = pd.read_excel('/content/3. 스마트스토어_상품_마케팅채널_2023-01-01_2023-03-23.xlsx')

df = pd.concat([df_1,df_2,df_3])
df.to_excel('ss_마케팅채널.xlsx', index=False)
```

```python
# '[사업자]' 단어가 들어가는 행들을 필터링하여 새로운 데이터 프레임 생성
filtered_df = df[df['상품명'].str.contains('\[사업자\]')]
sorted_df = filtered_df.sort_values(by='결제수(과거 14일간 기여도추정)', ascending=False)

# 차트 생성
plt.bar(sorted_df['상품명'], sorted_df['결제수(과거 14일간 기여도추정)'])
plt.xlabel('상품명')
plt.ylabel('결제수 (건)')
plt.title('[사업자] 결제 수')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sorted_df = filtered_df.sort_values(by='결제금액(과거 14일간 기여도추정)', ascending=False)
# 차트 생성
plt.bar(sorted_df['상품명'], sorted_df['결제금액(과거 14일간 기여도추정)'])
plt.xlabel('상품명')
plt.ylabel('결제금액')
plt.title('[사업자] 결제금액')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 상품명이 '[사업자] 클래식 블렌딩 / 1kg / 커피템플'인 제품만 필터링
filtered_df = df[df['상품명'] == '[사업자] 클래식 블렌딩 / 1kg / 커피템플']

# 채널명에 따른 결제 수를 그룹화하여 내림차순으로 정렬
grouped_df = filtered_df.groupby('채널명')['결제수(과거 14일간 기여도추정)'].sum().reset_index()
sorted_df = grouped_df.sort_values(by='결제수(과거 14일간 기여도추정)', ascending=False)

# 차트 생성
plt.bar(sorted_df['채널명'], sorted_df['결제수(과거 14일간 기여도추정)'])
plt.xlabel('채널명')
plt.ylabel('결제수')
plt.title('[사업자] 클래식 블렌딩 / 1kg / 커피템플 제품의 채널별 결제 수 (내림차순 정렬)')
plt.xticks(rotation=45, ha='right')

# 각 막대의 상단에 값을 표시
for index, value in enumerate(sorted_df['결제수(과거 14일간 기여도추정)']):
    plt.text(index, value, int(value), ha='center', va='bottom')

plt.show()
```

```python
# 채널명에 따른 결제 수를 그룹화하여 내림차순으로 정렬
grouped_df = df.groupby('채널명')['결제수(과거 14일간 기여도추정)'].sum().reset_index()
sorted_df = grouped_df.sort_values(by='결제수(과거 14일간 기여도추정)', ascending=False)

# 차트 생성
plt.bar(sorted_df['채널명'], sorted_df['결제수(과거 14일간 기여도추정)'])
plt.xlabel('채널명')
plt.ylabel('결제수')
plt.title('채널별 결제수')
plt.xticks(rotation=45, ha='right')

# 각 막대의 상단에 값을 표시
for index, value in enumerate(sorted_df['결제수(과거 14일간 기여도추정)']):
    plt.text(index, value, int(value), ha='center', va='bottom')

plt.show()
```

```python
# 채널명에 따른 결제 수를 그룹화하여 내림차순으로 정렬
grouped_df = df.groupby('채널명')['결제금액(과거 14일간 기여도추정)'].sum().reset_index()
sorted_df = grouped_df.sort_values(by='결제금액(과거 14일간 기여도추정)', ascending=False)

# 차트 생성
plt.bar(sorted_df['채널명'], sorted_df['결제금액(과거 14일간 기여도추정)'])
plt.xlabel('채널명')
plt.ylabel('결제수')
plt.title('채널별 결제금액')
plt.xticks(rotation=45, ha='right')


plt.show()
```
