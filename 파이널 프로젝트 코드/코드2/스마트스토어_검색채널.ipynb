```python
import koreanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_excel('/content/4. 스마트스토어_상품_검색채널_2022-01-01_2022-06-30.xlsx')
df_2 = pd.read_excel('/content/4. 스마트스토어_상품_검색채널_2022-07-01_2022-12-31.xlsx')
df_3 = pd.read_excel('/content/4. 스마트스토어_상품_검색채널_2023-01-01_2023-03-23.xlsx')

df = pd.concat([df_1,df_2,df_3])
df.to_excel('ss_검색채널.xlsx', index=False)
```

```python
df
```

```python
# 키워드 컬럼이 '커피템플'인 데이터만 필터링
filtered_df = (df[df['키워드'].str.contains(r'커피템플|템플|커피탬플|카페템플|커피템픅|커피텝플|탬플커피|coffeetemple|커피템츨|커피테플', regex=True)])

# '커피템플'인 데이터의 비율 계산
coffee_temple_count = len(filtered_df)
total_count = len(df)
ratio = coffee_temple_count / total_count

colors = ['#ff9999', 'lightgrey']
explode = (0.05, 0)
# 파이 차트 생성
plt.pie([ratio, 1-ratio], labels=['커피템플', '다른키워드'],  colors=colors, autopct='%1.1f%%', explode = explode)
plt.title('스마트스토어 유입 검색어')
plt.show()
```

```python
df['키워드'] = df['키워드'].str.replace(r'커피템플|템플|커피탬플|카페템플|커피템픅|커피텝플|탬플커피|coffeetemple|커피템츨|커피테플', '커피템플')
```

```python
# 키워드 필터링
not_contains_coffee_temple_df = df[~df['키워드'].str.contains(r'(검색어 없음)', regex=True)]
order_count = not_contains_coffee_temple_df.groupby('키워드')['결제수(과거 14일간 기여도추정)'].sum()

# 주문건수별로 내림차순 정렬
order_count_sorted = order_count.sort_values(ascending=False)

# 키워드별 색상 리스트
colors = ['#8B4513', '#654321', '#CD853F', '#D2691E', '#A0522D']

# 막대 차트 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(order_count_sorted.index, order_count_sorted.values, color=colors)
plt.xlabel('키워드')
plt.ylabel('주문건수')
plt.title('검색 키워드에 따른 주문건수')
plt.xticks(rotation=45, ha='right')

# 막대에 주문건수를 나타내는 텍스트 추가
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.show()
```
