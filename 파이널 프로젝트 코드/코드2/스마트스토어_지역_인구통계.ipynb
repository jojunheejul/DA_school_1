```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_excel('/content/7. 스마트스토어_상품_지역_2022-01-01_2022-06-30.xlsx')
df_2 = pd.read_excel('/content/7. 스마트스토어_상품_지역_2022-07-01_2022-12-31.xlsx')
df_3 = pd.read_excel('7. 스마트스토어_상품_지역_2023-01-01_2023-03-23.xlsx')

df_7 = pd.concat([df_1,df_2,df_3])
df_7.to_excel('ss_7.지역.xlsx', index=False)
```

```python
df_1 = pd.read_excel('/content/5. 스마트스토어_상품_인구통계_2022-01-01_2022-06-30.xlsx')
df_2 = pd.read_excel('/content/5. 스마트스토어_상품_인구통계_2022-07-01_2022-12-31.xlsx')
df_3 = pd.read_excel('/content/5. 스마트스토어_상품_인구통계_2023-01-01_2023-03-23.xlsx')

df_5 = pd.concat([df_1,df_2,df_3])
df_5.to_excel('ss_5.인구.xlsx', index=False)
```

```python
# 카테고리별 환불수량 계산
refund_counts = df_5.groupby('상품카테고리(대)')['환불수량'].sum()

# 전체 환불수량 계산
total_refund = refund_counts.sum()

# 퍼센트 계산
refund_percent = (refund_counts / total_refund) * 100

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.pie(refund_percent, labels=refund_percent.index, autopct='%.1f%%', colors=['skyblue', 'pink', 'orange', 'green'], startangle=90)
plt.title('상품 카테고리별 환불 비율', fontsize=16)
plt.axis('equal')  # 원형을 유지하기 위해 설정

plt.show()
```

```python
# 나이대별 환불수량 계산
refund_counts = df_5.groupby('나이')['환불수량'].sum()

# 전체 환불수량 계산
total_refund = refund_counts.sum()

# 퍼센트 계산
refund_percent = (refund_counts / total_refund) * 100

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.bar(refund_percent.index, refund_percent, color='skyblue')
plt.title('나이대별 환불 비율', fontsize=16)
plt.xlabel('나이대', fontsize=12)
plt.ylabel('환불 비율 (%)', fontsize=12)
plt.xticks(fontsize=10)  # x축 레이블의 폰트 크기 조정
plt.show()
```

```python
# 성별 별 환불건수 계산
refund_counts = df_5.groupby('성별')['환불수량'].sum()

# 전체 환불건수 계산
total_refund = refund_counts.sum()

# 퍼센트 계산
refund_percent = (refund_counts / total_refund) * 100

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.pie(refund_percent, labels=refund_percent.index, autopct='%.1f%%', colors=['gray', 'skyblue','pink'], startangle=90)
plt.title('성별에 따른 환불 비율', fontsize=16)
plt.axis('equal')
plt.subplots_adjust(top=1)
plt.show()
```

```python
df_5 = df_5.sort_values(by='결제금액', ascending=False)
# 나이대별 구매금액 그래프 그리기
plt.figure(figsize=(8, 6))
plt.bar(df_5['나이'], df_5['결제금액'])
plt.title('나이대별 구매금액', fontsize=16)
plt.xlabel('나이대', fontsize=12)
plt.ylabel('결제금액', fontsize=12)
plt.xticks(fontsize=10, rotation=45, ha='right')
plt.show()
```

```python
df_5 = df_5.sort_values(by='환불금액', ascending=False)
# 나이대별 환불금액 그래프 그리기
plt.figure(figsize=(8, 6))
plt.bar(df_5['나이'], df_5['환불금액'])
plt.title('나이대별 환불금액', fontsize=16)
plt.xlabel('나이대', fontsize=12)
plt.ylabel('환불금액', fontsize=12)
plt.xticks(fontsize=10, rotation=45, ha='right')
plt.show()
```

```python
# 전체 결제수량과 환불수량 계산
total_payments = df_5['결제상품수량'].sum()
total_refunds = df_5['환불수량'].sum()

# 환불수량의 비율 계산
refund_percent = (total_refunds / total_payments) * 100

# 그래프 그리기
plt.figure(figsize=(6, 6))
plt.pie([total_payments, total_refunds], labels=['결제상품수량', '환불수량'], autopct='%.1f%%', colors=['skyblue', 'pink'], startangle=90)
plt.title('결제수량 대비 환불수량 비율', fontsize=16)
plt.axis('equal')

plt.show()
```

```python
# 남성의 결제수량과 환불수량 계산
total_payments_male = df_5[df_5['성별'] == '남성']['결제상품수량'].sum()
total_refunds_male = df_5[df_5['성별'] == '남성']['환불수량'].sum()

# 여성의 결제수량과 환불수량 계산
total_payments_female = df_5[df_5['성별'] == '여성']['결제상품수량'].sum()
total_refunds_female = df_5[df_5['성별'] == '여성']['환불수량'].sum()

# 남성과 여성의 환불수량 비율 계산
refund_percent_male = (total_refunds_male / total_payments_male) * 100
refund_percent_female = (total_refunds_female / total_payments_female) * 100

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.pie([total_payments_male, total_refunds_male], labels=['결제', '환불'], autopct='%.1f%%', colors=['skyblue', 'pink'], startangle=90)
plt.title('남성 결제수량 대비 환불수량 비율', fontsize=16)
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.pie([total_payments_female, total_refunds_female], labels=['결제', '환불'], autopct='%.1f%%', colors=['skyblue', 'pink'], startangle=90)
plt.title('여성 결제수량 대비 환불수량 비율', fontsize=16)
plt.axis('equal')

plt.show()
```

```python
# 날짜를 연도별로 분리
df_1['연도'] =pd.to_datetime(df_1['날짜']).dt.year

# 2022년과 2023년 데이터 분리
df_2022 = df_1[df_1['연도'] == 2022]
df_2023 = df_1[df_1['연도'] == 2023]


# 2022년 시간대별 결제수 계산
payment_counts_by_time_2022 = df_2022.groupby(df_2022['시간대'])['결제당 결제금액'].sum()

# 2023년 시간대별 결제수 계산
payment_counts_by_time_2023 = df_2023.groupby(df_2023['시간대'])['결제당 결제금액'].sum()

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(payment_counts_by_time_2022.index, payment_counts_by_time_2022.values, marker='o', color='blue', label='2022년')
plt.plot(payment_counts_by_time_2023.index, payment_counts_by_time_2023.values, marker='o', color='orange', label='2023년')
plt.title('2022년과 2023년 시간대별 결제당 결제금액', fontsize=16)
plt.xlabel('시간대', fontsize=12)
plt.ylabel('결제당 결제금액', fontsize=12)
plt.xticks(rotation=0, ha='center')
plt.grid()
plt.legend()
plt.show()
```

```python
# 2022년 시간대별 결제수 계산
payment_counts_by_time_2022 = df_2022.groupby(df_2022['시간대'])['결제수'].sum()

# 2023년 시간대별 결제수 계산
payment_counts_by_time_2023 = df_2023.groupby(df_2023['시간대'])['결제수'].sum()

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(payment_counts_by_time_2022.index, payment_counts_by_time_2022.values, marker='o', color='blue', label='2022년')
plt.plot(payment_counts_by_time_2023.index, payment_counts_by_time_2023.values, marker='o', color='orange', label='2023년')
plt.title('2022년과 2023년 시간대별 결제수', fontsize=16)
plt.xlabel('시간대', fontsize=12)
plt.ylabel('결제수', fontsize=12)
plt.xticks(rotation=0, ha='center')
plt.grid()
plt.legend()
plt.show()
```
