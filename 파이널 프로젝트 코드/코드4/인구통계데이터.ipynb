```python

# Grouping by '채널명' (channel name) and calculating the count for each channel
grouped_data = sorted_da.groupby('채널명').size().reset_index(name='count')

# Calculate the percentage of each channel's count relative to the total
total_count = grouped_data['count'].sum()
grouped_data['퍼센트'] = grouped_data['count'] / total_count * 100

# Sorting the DataFrame by the count in descending order
sorted_grouped_data = grouped_data.sort_values(by='count', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
plt.bar(sorted_grouped_data['채널명'], sorted_grouped_data['퍼센트'])
plt.xlabel('채널명')
plt.ylabel('퍼센트(%)')
plt.title('채널별 빈도수 비율')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
