# 나의지갑관리DATA

```python
import pandas as pd
```

```python
wallet_df = pd.read_excel('/content/나의지갑관리DATA_20220601_20230430_map.xlsx')
wallet_df
```

```python
wallet_df.info()
```

```python
# 보기 쉽게 컬럼의 순서를 조정
new_columns = ['사입자명', '사입자아이디', '사입업체명', '지역명', '지역코드', '사입일', '등록일', '전잔', '입금액', '대납비', '당일잔금']
wallet_df = wallet_df.reindex(columns=new_columns)
wallet_df
```

```python
wallet_df['당일잔금_검증'] = wallet_df['전잔'] + wallet_df['입금액'] - wallet_df['대납비']
wallet_df
```

```python
중복제거_총사입업체명 = wallet_df['사입업체명'].unique()
중복제거_총사입업체명
```

```python
# 당일잔액은 초기 1달정도는 안 맞을 수 있음 -> 당일잔액이 맞춰진 거래만 추출
filtered_df = wallet_df[wallet_df['당일잔금'] == wallet_df['당일잔금_검증']]
filtered_df
```

```python
# 대납이 실제로 이루어진 거래만 추출
당일대납_df = filtered_df[(filtered_df['대납비'] != 0)]
당일대납_df
```

```python
당일대납_총사입업체명 = 당일대납_df['사입업체명'].unique()
당일대납_총사입업체명
```

```python
# 최종적으로 당일잔금이 (-)인 경우만 추출
minus_balance_df = 당일대납_df[당일대납_df['당일잔금'] < 0]
minus_balance_df
```

```python
# Dictionary mapping '사입업체명' to '지역명'
company_region_mapping = {
    'B1': '제주',
    'B2': '순천',
    'B3': '김포',
    'B4': '마산서울경기',
    'B5': '춘천',
    'B6': '의정부남양주',
    'B7': '강남제주',
    'B8': '남양주',
    'B9': '부산',
    'B10': '의정부',
    'B11': '목동',
    'B12': '대전',
    'B13': '춘천',
    'B14': '부산',
    'B15': '부산',
    'B16': '대구',
    'B17': '광주',
    'B18': '서울',
    'B22': '광주'
}

# 결과1과 결과2에서 '사입업체명'을 기준으로 그룹화하여 개수(count) 계산
result1_counts = 당일대납_df.groupby('사입업체명').size()
result2_counts = minus_balance_df.groupby('사입업체명').size()

# 결과2를 결과1에 맞춰 비율 계산
ratio_df = result2_counts / result1_counts
ratio_df = ratio_df.dropna()  # NaN 값 제거

# 데이터프레임 생성 및 지역명 매칭
df = pd.DataFrame({'사입업체명': ratio_df.index, '결과': ratio_df.values})
df['지역명'] = df['사입업체명'].map(company_region_mapping)

# 당일잔금_마이너스 컬럼 추가
df['당일잔금_마이너스'] = minus_balance_df.groupby('사입업체명').size().reindex(df['사입업체명']).fillna(0).astype(int).values

# 전체당일잔금 컬럼 추가
df['전체당일잔금'] = 당일대납_df.groupby('사입업체명').size().reindex(df['사입업체명']).fillna(0).astype(int).values

# 컬럼 순서 조정
df = df[['사입업체명', '지역명', '전체당일잔금', '당일잔금_마이너스', '결과']]

# 엑셀로 내보내기
df.to_excel('ratio_data.xlsx', index=False)
df
```

```python
# 마이너스 비율 높은 순서대로

# Dictionary mapping '사입업체명' to '지역명'
company_region_mapping = {
    'B1': '제주',
    'B2': '순천',
    'B3': '김포',
    'B4': '마산서울경기',
    'B5': '춘천',
    'B6': '의정부남양주',
    'B7': '강남제주',
    'B8': '남양주',
    'B9': '부산',
    'B10': '의정부',
    'B11': '목동',
    'B12': '대전',
    'B13': '춘천',
    'B14': '부산',
    'B15': '부산',
    'B16': '대구',
    'B17': '광주',
    'B18': '서울',
    'B22': '광주'
    # 필요한 경우 매핑 추가
}

# 결과1과 결과2에서 '사입업체명'을 기준으로 그룹화하여 개수(count) 계산
result1_counts = 당일대납_df.groupby('사입업체명').size()
result2_counts = minus_balance_df.groupby('사입업체명').size()

# 결과2를 결과1에 맞춰 비율 계산
ratio_df = result2_counts / result1_counts
ratio_df = ratio_df.dropna()  # NaN 값 제거

# 데이터프레임 생성 및 지역명 매칭
df = pd.DataFrame({'사입업체명': ratio_df.index, '결과': ratio_df.values})
df['지역명'] = df['사입업체명'].map(company_region_mapping)

# 당일잔금_마이너스 컬럼 추가
df['당일잔금_마이너스'] = minus_balance_df.groupby('사입업체명').size().reindex(df['사입업체명']).fillna(0).astype(int).values

# 전체당일잔금 컬럼 추가
df['전체당일잔금'] = 당일대납_df.groupby('사입업체명').size().reindex(df['사입업체명']).fillna(0).astype(int).values

# 비율을 기준으로 내림차순 정렬
df = df.sort_values(by='결과', ascending=False)

# 결과값을 소숫점 두 번째 자리까지로 잘라내기
df['결과'] = df['결과'].round(3)

# 컬럼 순서 조정
df = df[['사입업체명', '지역명', '전체당일잔금', '당일잔금_마이너스', '결과']]

# 엑셀로 내보내기
df.to_excel('ratio_data.xlsx', index=False)
df
```

```python
b1_data = df[df['사입업체명'] == 'B1']

# '당일잔금_마이너스'와 '전체당일잔금' 선택
minus_balance = b1_data['당일잔금_마이너스'].values[0]
total_balance = b1_data['전체당일잔금'].values[0]

# 비율 계산
ratio = minus_balance / total_balance
ratio = round(ratio * 100, 1)

# 파이 차트 그리기
sizes = [minus_balance, total_balance - minus_balance]
labels = ['당일잔금_마이너스', '당일잔금_플러스']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'fontsize':11, 'color': 'black'})
ax.axis('equal')  # 원형으로 설정

plt.title(f"B1(제주) 사입업체의 당일잔금 마이너스 비율", fontsize=14)

plt.subplots_adjust(top=1.4)

# 그래프 출력
plt.show()
```
